```python
#imports
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
# Create an empty list to collect article titles
titles = []

# Create an empty list to collect url of articles
links = []
```

```python
page = requests.get("https://www.barelaw.in/legal-drafts/", headers={"User-Agent": "XY"})

soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
req_h2_list = soup.find_all('h2', class_ = re.compile('post-item-title wi-post-title fox-post-title post-header-section size-normal')) 
```

```python
for title in req_h2_list:
    titles.append(title.text)
```

```python
titles
```

```python
req_a_list = soup.find_all('a', class_="readmore")
```

```python
for link in req_a_list:
    links.append(link.get('href'))
```

```python
links
```

```python
# Check the length of titles extracted
len(titles)
```

```python
# Check the length of the links extracted
len(links)
```

```python
# Create a dataframe from these collected lists of data
df = pd.DataFrame({"title":titles, "link":links})
```
